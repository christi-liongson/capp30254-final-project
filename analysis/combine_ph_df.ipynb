```python
MARSHALL_CSV = '../data/marshall_covid_cases.csv'

marshall_dtypes = {'name': str,
                   'total_staff_cases': 'Int64',
                   'total_prisoner_cases': 'Int64',
                   'total_staff_deaths': 'Int64',
                   'total_prisoner_deaths': 'Int64',
                   'as_of_date': str}

marshall = clean_data.import_csv(MARSHALL_CSV, marshall_dtypes)

marshall['as_of_date'] = pd.to_datetime(marshall['as_of_date'], format='%Y-%m-%d')
marshall['lower_name'] = marshall['name'].str.lower()

marshall.sort_values(by='as_of_date', inplace=True)

for state in list(marshall['name'].unique()):
    state_filter = marshall['name'] == state
    for col in marshall.select_dtypes(include='number').columns.to_list():
        marshall.loc[state_filter, col] = marshall.loc[state_filter, col].fillna(method='bfill')
        marshall.loc[state_filter, col] = marshall.loc[state_filter, col].fillna(method='ffill')

state_policies = marshall.merge(df[demographics], left_on='lower_name', right_on='state')
```
